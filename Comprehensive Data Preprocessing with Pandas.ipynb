```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon',
                              'pastrami', 'honey ham', 'nova lox'],
                     'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
  'bacon': 'pig',
  'pulled pork': 'pig',
  'pastrami': 'cow',
  'corned beef': 'cow',
  'honey ham': 'pig',
  'nova lox': 'salmon'
}
```

```python
lowercased=data['food'].str.lower()
lowercased
```

```python
#lowercased=data['food'].str.len()
```

```python
data['animal']=lowercased.map(meat_to_animal)
data
```

```python
data.groupby('animal')['ounces'].sum()   #dimension is animal and measure is ounces
```

```python
data
```

```python
data.replace('cow','buffalo')
```
