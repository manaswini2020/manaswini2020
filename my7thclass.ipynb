```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
    
chipo = pd.read_csv(url, sep = '\t')
chipo
```

```python
#item_price greater than $10
#chipo[chipo.item_price>10] it will not work bcz those values are string having $ sign
chipo.item_price.str.replace('$','')
```

```python
prices=[float(value[1:]) for value in chipo.item_price]
prices                            #I run it again so error showing
       
```

```python
chipo.item_price=prices
chipo
```

```python
chipo['unit_price']=chipo.item_price/chipo['quantity']
chipo
```

```python
#get rid of duplication
chipo_filtered = chipo.drop_duplicates('item_name')
chipo_filtered
```

```python
menu = chipo_filtered[['item_name','unit_price']]
menu.sort_values(by='unit_price',ascending=False).reset_index(drop=True) #we do nt want old order index so drop=true
```

```python
(menu.unit_price>10).sum()
```
